## Applying multinomial naive bayes from scratch

```python
#fit function
def fit(X_train,Y_train):#X_train and Y_train are numpy arrays
    result = {}
    classes = set(Y_train)
    result["all_Y_points"] = len(Y_train)
    for current_class in classes:
        result[current_class] = {}
        current_rows = (Y_train == current_class)
        Y_train_current = Y_train[current_rows]
        X_train_current = X_train[current_rows]
        result[current_class]["current_Y_points"] = len(Y_train_current)
        features = X_train.shape[1]
        #result[current_class]["total_X_points"] = 0
        sum_total_points = 0
        for wj in range(1,features+1):
            result[current_class][wj-1] = X_train_current[:,wj-1].sum()
            sum_total_points += X_train_current[:,wj-1].sum()
            #result[current_class]["total_X_points"] += X_train_current[:,wj-1].sum()
        result[current_class]["total_X_points"] =sum_total_points 
    return result
        
            
    
```

```python
# this gives us prob of each class 
def probability(x,dictionary,current_class):
    outcome = (np.log(dictionary[current_class]["current_Y_points"])) - np.log(dictionary["all_Y_points"])
    features = len(dictionary[current_class].keys())-2
    for wj in range(1,features+1):
        if x[wj-1] ==0 or x[wj-1]<2: # checking if that feature was present or not in that doc
            continue
        num = dictionary[current_class][wj-1] +1
        den = dictionary[current_class]["total_X_points"] + len(dictionary[current_class].keys())-2
        prob = np.log(num) - np.log(den)
        outcome += prob
    return outcome
        
```

```python
# this gives us class of a partiular x
def singlePointClass(x,dictionary):
    classes = dictionary.keys()
    best_prob = -10000
    predicted_class = " "
    first_one  = True
    for current_class in classes:
        if current_class == "all_Y_points":
            continue
        current_prob = probability(x,dictionary,current_class)
        if first_one or current_prob>best_prob:
            best_prob = current_prob
            predicted_class = current_class
            print(predicted_class)
        first_one = False
    return predicted_class
        
        
```

```python
# predict function

def predict(X_test,dictionary):
    Y_pred = []
    for x in X_test:
        class_predicted = singlePointClass(x,dictionary)
        Y_pred.append(class_predicted)
    return Y_pred
        
```
